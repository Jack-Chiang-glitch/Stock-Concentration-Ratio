```python
class EnterPosition:
    def __init__(self, ticker):
        self.ticker = ticker
        self.SCR_df = self.get_SCR_df()
        self.stock_price = self.get_stock_price()
        self.enter_position = self.get_EnterPosition()
        

    def get_SCR_df(self):
        file_path = os.path.join('SCR', f"{self.ticker}.csv")
        df = pd.read_csv(file_path, index_col=0)  # 假設索引是 CSV 中的第一列
        df.replace('nan', np.nan, inplace=True)
        df.dropna(inplace=True)  # 這裡進行就地刪除
        return df

    def get_stock_price(self):
        stock_data = yf.download(f"{self.ticker}.TW", start="2024-01-01", end="a-01-16")
        if stock_data.empty:
            stock_data = yf.download(f"{self.ticker}.TWO", start="2024-01-01", end="2025-01-16")
        #df.index
        stock_data = stock_data.loc[self.SCR_df.index]
        stock_price = stock_data['Adj Close']
        # 確保 stock_price.index 也是 datetime 類型
        stock_price.index = pd.to_datetime(stock_price.index)
        return stock_price

    def get_EnterPosition(self):
        enter_position = []
        for date in self.SCR_df.index:
            single_day_date = self.SCR_df.loc[date]
            SCR1, SCR5, SCR60, SCR120 = \
                single_day_date['1'], single_day_date['5'], single_day_date['60'], single_day_date['120']

            # 黃金交叉
            
            
            # 1.多頭排列：1日>5日>60日>120日 
            #condition = (SCR1>SCR5) & (SCR5>SCR60) & (SCR60>120)
            
            # 1.多頭排列：1日>5日>60日
            #condition = (SCR1>SCR5) & (SCR5>SCR60)

            #4-2反種型態：5日>60日&120日
            #condition = (SCR5>SCR60) & (SCR5>SCR120)

            #4-3反種型態：60日>5日>120日
            #condition = (SCR60>SCR5) & (SCR5>SCR120)

            #2.空頭排列：1日<5日<60日<120日
            #condition = (SCR1<SCR5) & (SCR5<SCR60) & (SCR60<120)

            #3.乖離型態：「1&5日」與120日數值相差4%以上以及與60日數值相差2%
            """
            condition = (
                (abs(SCR1 - SCR120) / SCR120 >= 0.04) &  # 1日與120日數值相差4%以上
                (abs(SCR1 - SCR60) / SCR60 >= 0.02) &   # 1日與60日數值相差2%以上
                (abs(SCR5 - SCR120) / SCR120 >= 0.04) &  # 5日與120日數值相差4%以上
                (abs(SCR5 - SCR60) / SCR60 >= 0.02)     # 5日與60日數值相差2%以上
            )
            """

            
            if condition: 
                enter_position.append(date)
                
        # 確保 enter_position 是 datetime 類型
        enter_position = pd.to_datetime(enter_position)
        return enter_position

    

    def plot_SCR_price(self):
        # 確保 SCR_df 的 index 是 datetime 類型
        self.SCR_df.index = pd.to_datetime(self.SCR_df.index, errors='coerce')  # 'coerce' 會將無法轉換的值設為 NaT
        
        # 創建圖形和第一個Y軸
        fig, ax1 = plt.subplots(figsize=(10, 4.8))
        
        # 繪製股價圖
        ax1.plot(self.stock_price.index, self.stock_price, label='Stock Price', color='blue')
        
        # 設定第一個Y軸標題和標籤
        ax1.set_title(f'Stock Price with Enter Positions ({self.ticker})')
        ax1.set_xlabel('Date')
        ax1.set_ylabel('Stock Price', color='blue')
        ax1.tick_params(axis='y', labelcolor='blue')
        
        # 創建第二個Y軸
        ax2 = ax1.twinx()
        
        # 繪製 SCR1, SCR5, SCR60, SCR120 在第二個Y軸
        ax2.plot(self.SCR_df.index, self.SCR_df['1'], label='SCR1', color='red', linestyle='--')
        ax2.plot(self.SCR_df.index, self.SCR_df['5'], label='SCR5', color='green', linestyle='--')
        ax2.plot(self.SCR_df.index, self.SCR_df['60'], label='SCR60', color='orange', linestyle='--')
        ax2.plot(self.SCR_df.index, self.SCR_df['120'], label='SCR120', color='purple', linestyle='--')
        
        # 設定第二個Y軸標題和標籤
        ax2.set_ylabel('SCR Values', color='black')
        ax2.tick_params(axis='y', labelcolor='black')
        
        # 添加圖例
        ax1.legend(loc='upper left')
        ax2.legend(loc='upper right')
        
        # 顯示圖表
        ax1.grid(True)
        plt.xticks(rotation=10)
        
        # 儲存圖像到指定位置
        plt.savefig('/path/to/your/directory/SCR_price_plot.png', dpi=300, bbox_inches='tight')  # 更改路徑和文件名
        
        # 顯示圖表
        plt.show()


    def plot_SCR_return(self):
        # 確保 SCR_df 的 index 是 datetime 類型
        self.SCR_df.index = pd.to_datetime(self.SCR_df.index, errors='coerce')  # 'coerce' 會將無法轉換的值設為 NaT
        
        # 創建圖形和第一個Y軸
        fig, ax1 = plt.subplots(figsize=(10, 4.8))
        
        # 繪製股價圖
        returns = self.stock_price.pct_change()
        ax1.plot(returns.index, returns, label='Stock Return', color='blue')
        
        # 設定第一個Y軸標題和標籤
        ax1.set_title(f'Stock Return with Enter Positions ({self.ticker})')
        ax1.set_xlabel('Date')
        ax1.set_ylabel('Stock Return', color='blue')
        ax1.tick_params(axis='y', labelcolor='blue')
        
        # 創建第二個Y軸
        ax2 = ax1.twinx()
        
        # 繪製 SCR1, SCR5, SCR60, SCR120 在第二個Y軸
        ax2.plot(self.SCR_df.index, self.SCR_df['1'], label='SCR1', color='red', linestyle='--')
        ax2.plot(self.SCR_df.index, self.SCR_df['5'], label='SCR5', color='green', linestyle='--')
        ax2.plot(self.SCR_df.index, self.SCR_df['60'], label='SCR60', color='orange', linestyle='--')
        ax2.plot(self.SCR_df.index, self.SCR_df['120'], label='SCR120', color='purple', linestyle='--')
        
        # 設定第二個Y軸標題和標籤
        ax2.set_ylabel('SCR Values', color='black')
        ax2.tick_params(axis='y', labelcolor='black')
        
        # 添加圖例
        ax1.legend(loc='upper left')
        ax2.legend(loc='upper right')
        
        # 顯示圖表
        ax1.grid(True)
        plt.xticks(rotation=10)
        plt.savefig(f'報酬率SCR圖/{self.ticker}報酬率SCR.png', dpi=300, bbox_inches='tight')
        #plt.show()
    
    def plot_enter_position(self):
        # 繪製股票價格圖
        plt.figure(figsize=(10, 4.8))
        plt.plot(self.stock_price.index, self.stock_price, label='Stock Price', color='blue')
        
        # 在進場日期上標註紅點
        for i, date in enumerate(self.enter_position):
            if date in self.stock_price.index:  # 確保進場日期存在於 stock_price 中
                plt.scatter(date, self.stock_price.loc[date], s = 30, color='red', 
                            label='Enter Position' if i == 0 else "")  # 只在第一個標註時設置 label
        
        # 設定圖表標題和標籤
        plt.title(f'Stock Return with Enter Positions({self.ticker})')
        plt.xlabel('Date')
        plt.ylabel('Price')
        plt.xticks(rotation=10)
        plt.legend()
        
        # 顯示圖表
        plt.grid(True)
        plt.show()
    
   
#ticker_list = []
for ticker in ticker_list:
    try:
        E = EnterPosition(ticker)
        E.plot_enter_position()
    
        #E.plot_SCR_price()
        #E.plot_SCR_return()
    except:
        print(ticker)
    
```

# 股價SCR圖 or 報酬率SCR圖 轉成HTML

```python
import plotly.graph_objects as go
import base64
import plotly.io as pio

#mode = '股價'
mode = '報酬率'

def SaveImgAsHtml(ticker):
    # 讀取圖片並轉換為 Base64 編碼
    image_path = f'{mode}SCR圖/{ticker}{mode}SCR.png'  # 請替換為您圖片的路徑
    with open(image_path, "rb") as image_file:
        encoded_image = base64.b64encode(image_file.read()).decode()
    
    # 創建 Plotly 圖表
    fig = go.Figure()
    
    # 設置圖片作為背景，圖片會隨著圖表的縮放進行調整
    fig.update_layout(
        autosize=True,  # 開啟自動調整大小
        images=[dict(
            source=f"data:image/png;base64,{encoded_image}",
            x=0,  # X 軸位置，左側
            y=1,  # Y 軸位置，頂部
            xanchor="left",  # 錨點設置為左邊
            yanchor="top",  # 錨點設置為頂部
            sizex=1,  # 這裡設定為 1，這樣圖片會根據圖表自適應大小
            sizey=1,  # 這裡設定為 1，這樣圖片會根據圖表自適應大小
            opacity=0.5,
            layer="below"  # 背景圖片
        )],
        # 設定網格線顏色
        xaxis=dict(
            gridcolor='brown'  # X 軸網格線顏色
        ),
        yaxis=dict(
            gridcolor='brown'  # Y 軸網格線顏色
        )
    )
    
    # 儲存為 HTML 文件
    html_file_path = f'{mode}SCR圖html/{ticker}{mode}SCR.html'
    pio.write_html(fig, html_file_path)
    
    print(f"HTML 文件已儲存至 {html_file_path}")

for ticker in ticker_list:
    try:
        SaveImgAsHtml(ticker)
    except:
        print(f"{ticker} save failed")
```
